```python
dataset=pd.read_csv('dataset.csv')
```

```python
print(dataset.shape)
print(dataset.head(5))
```

```python
import matplotlib.pyplot as plt
plt.scatter(dataset.area,dataset.price,color='magenta',marker='*')
plt.show
```

```python
X=dataset.drop('price',axis=1)
X
```

```python
Y=dataset.price
Y
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(X,Y)
```

```python
x=int(input("enter the area whose price need to be predicted"))
y=[[x]]
PricePredictor=model.predict(y)
print(PricePredictor)
```

```python
m=model.coef_
m
```

```python
b=model.intercept_
b
```

```python
Pr=m*x+b
print("Price for the area is: ",Pr)
```
